#### Create a swarmplot using a melted dataframe of all your numeric variables. The x-axis should be the variable name, the y-axis the measure. Add another dimension using color to represent species. Document takeaways from this visualization.

```python
melted_train = pd.melt(train, id_vars='species', var_name='measure', value_name= 'value')
sns.swarmplot(data=melted_train, x='measure', y='value', hue='species')
```

#### Ask a specific question of the data, such as: is the sepal area signficantly different in virginica compared to setosa? Answer the question through both a plot and using a mann-whitney or t-test. If you use a t-test, be sure assumptions are met (independence, normality, equal variance).

```python
train['area_sepal'] = train.sepal_length * train.sepal_width
sevi_df = train[train['species'] != 2]
sns.boxplot(data=sevi_df, x=sevi_df['species'], y=sevi_df['area_sepal'])
```

```python
stats.mannwhitneyu(sevi_df[sevi_df.species == 0].area_sepal, 
                   sevi_df[sevi_df.species == 1].area_sepal)
```

#### Document takeaways and any actions.
- Sepal area is statistically significantly different with confidence level 95%
- The p-value is not as miniscule as petal length and width for determining species type
